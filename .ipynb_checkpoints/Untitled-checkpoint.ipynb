```python
import os

def count_files_and_lines(directory):
    file_counts = {}
    total_lines = 0
    
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_extension = os.path.splitext(file)[-1].lower()
            file_size = os.path.getsize(file_path)
            file_lines = count_lines(file_path)
            
            if file_extension not in file_counts:
                file_counts[file_extension] = {"count": 1, "lines": file_lines, "size": file_size}
            else:
                file_counts[file_extension]["count"] += 1
                file_counts[file_extension]["lines"] += file_lines
                file_counts[file_extension]["size"] += file_size
                
            total_lines += file_lines
            
    return file_counts, total_lines

def count_lines(file_path):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        return sum(1 for line in file)

def main():
    directory = "."  # 当前文件夹
    file_counts, total_lines = count_files_and_lines(directory)
    
    print("文件类型统计:")
    for extension, stats in file_counts.items():
        print(f"{extension}: {stats['count']} 个文件, {stats['lines']} 行代码, 总大小 {stats['size']} bytes.")
    
    print("\n总行数:", total_lines)

if __name__ == "__main__":
    main()
```

```python
import os

def count_xml_files(directory):
    xml_files = []
    
    for root, dirs, files in os.walk(directory):
        if ".idea" in dirs:
            dirs.remove(".idea")  # 忽略 .idea 文件夹
        for file in files:
            if file.endswith(".xml"):
                xml_files.append(os.path.join(root, file))
                
    return xml_files

def main():
    directory = "."  # 当前文件夹
    xml_files = count_xml_files(directory)
    
    print("所有的 XML 文件:")
    for xml_file in xml_files:
        print(xml_file)

if __name__ == "__main__":
    main()
```
